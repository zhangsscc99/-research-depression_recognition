```python
base_model = ResNet50(weights='imagenet', include_top=False)

# Add a new top layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
#x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='linear')(x)

# Create a new model
model = Model(inputs=base_model.input, outputs=x)
```
